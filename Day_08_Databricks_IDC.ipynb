```python
%sql
-- # Create catalog & schemas 

CREATE catalog IF NOT EXISTS ecommerce_IDC;
```

```python
%sql
use catalog ecommerce_idc;
CREATE SCHEMA IF NOT EXISTS bronze;
CREATE SCHEMA IF NOT EXISTS silver;
CREATE SCHEMA IF NOT EXISTS gold;
```

```python
%sql
SHOW SCHEMAS;
```

```python
%sql

CREATE TABLE IF NOT EXISTS ecommerce_idc.bronze.events (
    event_time TIMESTAMP,
    user_id STRING,
    product_id STRING,
    price DOUBLE
)
USING DELTA;

CREATE TABLE IF NOT EXISTS ecommerce_idc.silver.events (
    event_time TIMESTAMP,
    user_id STRING,
    product_id STRING,
    price DOUBLE
)
USING DELTA;

CREATE TABLE IF NOT EXISTS ecommerce_idc.gold.products (
    product_id STRING,
    product_name STRING,
    revenue DOUBLE,
    purchases INT
)
USING DELTA;
```
